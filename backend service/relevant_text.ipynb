#### Assigning the labels to the rows

```python
def conditions(s):
    if 0.002 <= s["height"] <= 0.006:
        return "caption"
    elif 0.006 <= s["height"] <= 0.035:
        return "body"
    elif 0.035 <= s["height"] <= 0.3:
        return "header"
    elif 0.3 <= s["height"] <= 0.9:
        return "other"
```

```python
df["label"] = df.apply(conditions, axis=1)
```
